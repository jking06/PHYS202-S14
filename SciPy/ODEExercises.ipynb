### Question 1

For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

**(a)** See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, $θ_0$, spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?


```python
from scipy.integrate import odeint
```

```python
t = np.arange(0.,10.+0.01,0.01)

def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

plt.figure(figsize=(8,6), dpi=80)
plt.plot(t,odeint(deriv,np.array([0,0]),t)[:,0],'r-',label="0,0")
plt.plot(t,odeint(deriv,np.array([np.pi/4.,0]),t)[:,0],'b-',label="Pi/4,0")
plt.plot(t,odeint(deriv,np.array([np.pi/2.,0]),t)[:,0],'g-',label="Pi/2,0")
plt.plot(t,odeint(deriv,np.array([3./4.*np.pi,0]),t)[:,0],'k-',label="3Pi/4,0")
plt.plot(t,odeint(deriv,np.array([np.pi,0]),t)[:,0],'y-',label="Pi,0")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
#plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

At angles 0 and Pi, the pendulum is at equilibrium: straight up and down, respectively. Between $\pi/4$ and $3\pi/4$, the pendulum's ampilitude and wavelength increase.


**(b)** Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for $θ_0$ = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python
plt.figure(figsize=(8,6), dpi=80)
plt.plot(t,odeint(deriv,np.array([0.95*np.pi,0]),t)[:,0],'r-',label="Theta-t")
plt.plot(t,odeint(deriv,np.array([0.95*np.pi,0]),t)[:,1],'b-',label="Omega-t")
plt.plot(odeint(deriv,np.array([0.95*np.pi,0]),t)[:,1], odeint(deriv,np.array([0.95*np.pi,0]),t)[:,0], 'g-',label="Omega-Theta")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
#plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```
